```python
import sympy as sp

# Transition matrix
P = sp.Matrix([
    [0.2, 0.3, 0.1, 0.2, 0.2],
    [0.1, 0.4, 0.2, 0.2, 0.1],
    [0.3, 0.2, 0.2, 0.1, 0.2],
    [0.25, 0.15, 0.25, 0.2, 0.15],
    [0.2, 0.2, 0.2, 0.1, 0.3]
])

# Unknown stationary probabilities
pi1, pi2, pi3, pi4, pi5 = sp.symbols(
    'pi1 pi2 pi3 pi4 pi5'
)

# Row vector
pi = sp.Matrix([[pi1, pi2, pi3, pi4, pi5]])

# Stationary equations
equations = list(pi * P - pi)

# Normalization condition
equations.append(
    pi1 + pi2 + pi3 + pi4 + pi5 - 1
)

# Solve system
solution = sp.solve(
    equations,
    [pi1, pi2, pi3, pi4, pi5]
)

print(solution)
```
