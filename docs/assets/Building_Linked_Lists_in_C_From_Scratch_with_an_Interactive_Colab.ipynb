## 8. Two connected nodes

```python
%%bash
cat << 'EOF' > main.c
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

typedef struct ll_text {
    char *text;
    struct ll_text *next;
} ll_text;

int main() {
    ll_text *a = malloc(sizeof(ll_text));
    ll_text *b = malloc(sizeof(ll_text));

    a->text = strdup("hello");
    a->next = b;

    b->text = strdup("world");
    b->next = NULL;

    printf("%s -> %s -> NULL\n", a->text, a->next->text);

    free(a->text);
    free(b->text);
    free(b);
    free(a);

    return 0;
}
EOF

gcc main.c -o main
./main
```

## 9. Traversing a list

```python
%%bash
cat << 'EOF' > main.c
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

typedef struct ll_text {
    char *text;
    struct ll_text *next;
} ll_text;

int main() {
    ll_text *a = malloc(sizeof(ll_text));
    ll_text *b = malloc(sizeof(ll_text));
    ll_text *c = malloc(sizeof(ll_text));

    a->text = strdup("A");
    a->next = b;

    b->text = strdup("B");
    b->next = c;

    c->text = strdup("C");
    c->next = NULL;

    ll_text *curr = a;

    while (curr != NULL) {
        printf("%s -> ", curr->text);
        curr = curr->next;
    }

    printf("NULL\n");

    return 0;
}
EOF

gcc main.c -o main
./main
```

## 11. Original join_lists example

```python
%%bash
cat << 'EOF' > linked_list.c
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

typedef struct ll_text {
    char *text;
    struct ll_text *next;
} ll_text;

ll_text *join_lists(ll_text *list1, ll_text *list2) {
    ll_text *newlist = malloc(sizeof(ll_text));

    newlist->text = strdup(list1->text);
    newlist->next = list2;

    return newlist;
}

int main() {
    ll_text *list1 = malloc(sizeof(ll_text));
    list1->text = strdup("hello");
    list1->next = NULL;

    ll_text *list2 = malloc(sizeof(ll_text));
    list2->text = strdup("world");
    list2->next = NULL;

    ll_text *list3 = join_lists(list1, list2);

    ll_text *curr = list3;

    while (curr != NULL) {
        printf("%s -> ", curr->text);
        curr = curr->next;
    }

    printf("NULL\n");

    return 0;
}
EOF

gcc linked_list.c -o linked_list
./linked_list
```

## 12. Important exam idea

`join_lists` only copies ONE node.

It does NOT copy the whole first list.


## 13. Appending list2 to end of list1

```python
%%bash
cat << 'EOF' > main.c
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

typedef struct ll_text {
    char *text;

    struct ll_text *next;
} ll_text;

ll_text *make_node(char *word) {
    ll_text *node = malloc(sizeof(ll_text));
    node->text = strdup(word);
    node->next = NULL;
    return node;
}

ll_text *append_lists(ll_text *list1, ll_text *list2) {
    ll_text *curr = list1;

    while (curr->next != NULL) {
        curr = curr->next;
    }

    curr->next = list2;

    return list1;
}

void print_list(ll_text *head) {
    while (head != NULL) {
        printf("%s -> ", head->text);
        head = head->next;
    }

    printf("NULL\n");
}

int main() {
    ll_text *a = make_node("A");
    ll_text *b = make_node("B");
    ll_text *x = make_node("X");

    a->next = b;


    ll_text *joined = append_lists(a, x);

    print_list(joined);

    return 0;
}
EOF

gcc main.c -o main
./main
```

## 14. Common bug: forgetting NULL

```python
%%bash
cat << 'EOF' > main.c
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

typedef struct ll_text {
    char *text;
    struct ll_text *next;
} ll_text;

int main() {
    ll_text *node = malloc(sizeof(ll_text));

    node->text = strdup("hello");

    // IMPORTANT
    node->next = NULL;

    if (node->next == NULL) {
        printf("safe end of list\n");
    }

    free(node->text);
    free(node);

    return 0;
}
EOF

gcc main.c -o main
./main
```
