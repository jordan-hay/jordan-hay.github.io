```r
devtools::install_github("FinYang/tsdl")
library(tsdl)
print(tsdl)
```

```r
subject <- "Agriculture"
frequency <- 12
example <- 2
ts <- subset(tsdl, frequency, subject)
ts <- ts[[example]]
ts
```

```r
time_points <- time(ts)
time_points
```

```r
periods = cycle(ts)
periods
```

```r
years <- floor(time_points)
years
```

```r
attributes=attributes(ts)
attributes
```

```r
plot(ts,main = attributes$description)
```

```r
write.csv(ts, "data.csv", row.names = FALSE)
```
